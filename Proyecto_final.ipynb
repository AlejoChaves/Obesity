**Algoritmos de aprendizaje automático para análisis y predicción de datos relacionados con Obesidad**

Según la Organización Mundial de la Salud (OMS), desde 1975 casi que se ha triplicado la obesidad a nivel mundial y aunque en el 2020 fue la causa de muerte de aproximadamente 4 millones de personas, esta efermedad se puede prevenir. [1][2]

La OMS define el sobrepeso y la obesidad "como una acumulación anormal o excesiva de grasa que puede ser perjudicial para la salud". Usualmente se diagnostica a través de examenes físicos desde medicina general, en los cuales se indaga sobre los antecedentes familiares, hábitos de consumo, actividad física, entre otros factores de riesgo asociados al comportamiento y a las condiciones sociales y del entorno; adicionalmente, se realizan mediciones de peso y talla, se calcula el Índice de Masa Corporal (IMC) y se mide el perímetro abdominal. 

El Índice de Masa Corporal es un indicador que relaciona el peso y la talla del paciente, este se calcula dividiendo el peso corporal en kilogramos por la estatura al cuadrado, en metros. Un IMC superior a 25 indica sobrepeso y un IMC igual o mayor a 30 se considera obesidad.

La obesidad hace parte del grupo de enfermedades no transmisibles (ENT), las cuales suelen ser crónicas o de larga duración y generalmente surgen a partir de la combinación de factores genéticos, fisiológicos y comportamentales. Un IMC elevado es un factor que incrementa el riesgo de desarrollar otras ENT como enfermedades cardiovasculares, osteoarticulares, diabetes y algunos tipos de cáncer (endometrio, mama, ovarios, próstata, hígado, vesícula biliar, riñones y cólon). [1][3]

#4. Explicación del Balanceo con el filtro Smote

Un dataset se encuentra desequilibrado cuando las categorías de clasificación (target) no están representadas aproximadamente por igual; [5] en el caso del dataset en estudio la categoría de peso normal es predominante, respecto a las categorías de bajo peso, sobrepeso u obesidad. Uno de los problemas observados en la clasificación en datasets desequilibrados, es la posibilidad de clasificar erróneamente un registro, según Chawla, se conoce que "el costo de clasificar erróneamente un ejemplo anormal o interesante como un ejemplo normal, suele ser mucho mayor que el costo del error inverso" y propone que una buena forma de tratar este tipo de problemas es mediante una combinación de submuestreo de la clase mayoritaria con el sobremuestreo de la clase minoritaria a través de la generación de datos sintéticos [6]

En el trabajo de Mendoza y De La Hoz emplearon el filtro SMOTE (Synthetic Minority Over-sampling Technique) sugerido por Chawla, para lograr el balanceo de los datos. El filtro se aplica aleatorizando la clase minoritaria y con la técnica de K vecinos más cercanos se crean los nuevos datos y se guardan en un arreglo. 




Código para aplicar el filtro SMOTE: [6]


Algorithm SMOTE(T, N, k)
Input: Number of minority class samples T; Amount of SMOTE N%; Number of nearest
neighbors k
Output: (N/100) * T synthetic minority class samples
1. (∗ If N is less than 100%, randomize the minority class samples as only a random
percent of them will be SMOTEd. ∗)
2. if N < 100
3. then Randomize the T minority class samples
4. T = (N/100) ∗ T
5. N = 100
6. endif
7. N = (int)(N/100) (∗ The amount of SMOTE is assumed to be in integral multiples of
100. ∗)
8. k = Number of nearest neighbors
9. numattrs = Number of attributes
10. Sample[ ][ ]: array for original minority class samples
11. newindex: keeps a count of number of synthetic samples generated, initialized to 0
12. Synthetic[ ][ ]: array for synthetic samples
(∗ Compute k nearest neighbors for each minority class sample only. ∗)
13. for i ← 1 to T
14. Compute k nearest neighbors for i, and save the indices in the nnarray
15. Populate(N, i, nnarray)
16. endfor
Populate(N, i, nnarray) (∗ Function to generate the synthetic samples. ∗)
17. while N 6= 0
18. Choose a random number between 1 and k, call it nn. This step chooses one of
the k nearest neighbors of i.
19. for attr ← 1 to numattrs
20. Compute: dif = Sample[nnarray[nn]][attr] − Sample[i][attr]
21. Compute: gap = random number between 0 and 1
22. Synthetic[newindex][attr] = Sample[i][attr] + gap ∗ dif
23. endfor
24. newindex++
25. N = N − 1
26. endwhile
27. return (∗ End of Populate. ∗)
End of Pseudo-Code.

Teniendo en cuenta que el conjunto de datos presenta medidas que permiten conocer el nivel de obesidad de una persona, el problema se enfoca desde el marco del aprendizaje supervisado.
```python
models=[reg_lin, reg_SVR, reg_bg, reg_boost]
model_names=['KNeighborsRegressor','SVR','DecisionTreeRegressor','BaggingRegressor','AdaBoostRegressor']

for i, clf in enumerate(models):
  clf = clf.fit(X_train, y_train)
  print(model_names[i]+" training score :", clf.score(X_train,y_train))
  print(model_names[i]+" test score :", clf.score(X_test,y_test))
```

**MSE ERROR CUADRATICO MEDIO**

```python
from sklearn.metrics import mean_squared_error

for i, clf in enumerate(models):
  clf = clf.fit(X_train, y_train)
  y_train_pred=clf.predict(X_train)
  y_test_pred=clf.predict(X_test)
  print(model_names[i]+" train mse:", mean_squared_error(y_train, y_train_pred))
  print(model_names[i]+" test mse:", mean_squared_error(y_test, y_test_pred))
```

**Métrica MAE (Error absoluto medio)**

```python
from sklearn.metrics import mean_absolute_error

for i, clf in enumerate(models):
  clf = clf.fit(X_train, y_train)
  y_train_pred=clf.predict(X_train)
  y_test_pred=clf.predict(X_test)
  print(model_names[i]+" train mae:", mean_absolute_error(y_train, y_train_pred))
  print(model_names[i]+" test mae:", mean_absolute_error(y_test, y_test_pred))
```

**Métrica R2 ajustado**

```python
from sklearn.metrics import r2_score

def r2_score_adj(X,y,y_pred):
   r2=r2_score(y, y_pred)
   n,k=np.shape(X)
   adj_r2= 1 - ( (1-r2)*(n-1) / (n-k-1) )
   return adj_r2

for i, clf in enumerate(models):
  clf = clf.fit(X_train, y_train)
  y_train_pred=clf.predict(X_train)
  y_test_pred=clf.predict(X_test)
  print(model_names[i]+" train R^2:", r2_score(y_train, y_train_pred))
  print(model_names[i]+" test R^2:", r2_score(y_test, y_test_pred))
  print(model_names[i]+" train R^2 adjusted:", r2_score_adj(X_train, y_train, y_train_pred))
  print(model_names[i]+" test R^2 adjusted:", r2_score_adj(X_test, y_test, y_test_pred))
```
